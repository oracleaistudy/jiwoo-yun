### 지도 학습과 비지도 학습

- 지도 학습 : 알고리즘이 정답을 맞히는 것을 학습, 데이터(입력)와 정답(타깃)이 필요함
- 비지도 학습 : 타깃 없이 입력 데이터만 사용

```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

fish_data = [[l, w] for l, w in zip(fish_length, fish_weight)]
fish_target = [1]*35 + [0]*14  # 도마 35마리, 빙어 14마리

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

### 훈련 세트와 테스트 세트 (머신 러닝의 정확한 평가를 위해 따로 준비되어야 함)

- 훈련 세트 : 훈련에 사용되는 데이터
- 테스트 세트 : 평가에 사용하는 데이터

```python
print(fish_data[4])   # 인덱스 설정, 다섯 번쨰 샘플의 인덱스는 4
print(fish_data[0:5]) # 0부터 4까지 출력
print(fish_data[:5])  # 앞자리가 비면 처음부터 출력
print(fish_data[44:]) # 뒷자리가 비면 끝까지 출력

train_input = fish_data[:35]     # 훈련 세트로 입력값 중 0부터 34번째까지 사용
train_target = fish_target[:35]  # 훈련 세트로 타깃값 중 0부터 34번째까지 사용
test_input = fish_data[35:]      # 테스트 세트로 입력값 중 35번째부터 마지막 인덱스까지 사용
test_target = fish_target[35:]   # 테스트 세트로 타킷값 중 35번째부터 마지막 인덱스까지 사용

kn.fit(train_input, train_target)
kn.score(test_input, test_target)  # 샘플링 편향으로 인해 결과가 0으로 나옴
```

### 넘파이 (Numpy)

파이썬의 대표적인 배열 라이브러리

고차원의 배열을 손쉽게 만들고 조작할 수 있는 간편한 도구 제공

```python
import numpy as np

input_arr = np.array(fish_data)  # array() 함수를 사용하여 파이썬 리스트를 넘파이 배열로 바꿈
target_arr = np.array(fish_target)

print(input_arr)
print(input_arr.shape)  # (샘플 수, 특성 수) 출력
```

### arange() 함수

넘파이 arange() 함수를 사용하면 0에서부터 48까지 1씩 증가하는 인덱스를 만들 수 있음 


### random.seed 지정

넘파이에서 무작위 결과를 만드는 함수들은 실행할 때마다 다른 결과를 만듦 

일정한 결과를 얻으려면 초기에 random.seed를 지정하면 됨

<< 일정한 결과가 나와야하는 이유 >>
1. 재현성
2. 디버깅
3. 공정한 모델 비교
4. 연구·보고 정확성

```python
np.random.seed(42)
index = np.arange(49)
np.random.shuffle(index)

print(index)
```

```python
print(input_arr[[1,3]])

train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

print(input_arr[13], train_input[0])

test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]

# 산점도를 이용하여 훈련 세트와 테스트 세트에 도미와 빙어가 잘 섞였는지 확인
import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(test_input[:,0], test_input[:,1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 두 번째 머신러닝 프로그램

```python
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
kn.predict(test_input)
test_target
```

## CH 02-2 데이터 전처리

```python
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
import numpy as np
np.column_stack(([1,2,3],[4,5,6]))  # 전달받은 리스트를 일렬로 세운 후 차례대로 연결

fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data[:5])
```

```python
print(np.ones(5))

fish_target = np.concatenate((np.ones(35), np.zeros(14)))
print(fish_target)
```

### 사이킷런으로 훈련 세트와 테스트 세트 나누기
* train_test_split()

전달되는 리스트나 배열을 비율에 맞게 훈련 세트와 테스트 세트로 나눠줌

사이킷런의 model_selection 모듈 아래 있고 다음과 같이 임포트함

자체적으로 랜덤 시드를 지정할 수 있는 random_state 매개변수가 있음

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state=42)

print(train_input.shape, test_input.shape)
print(train_target.shape, test_target.shape)
print(test_target)

# stratify 매개변수에 타킷 데이터를 전달하여 클래스 비율에 맞게 데이터 나누기
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, stratify=fish_target, random_state=42)
print(test_target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)

print(kn.predict([[25, 150]]))  # 사이킷런에선 [[]]는 2차원 형태의 데이터를 위함

import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
distances, indexes = kn.kneighbors([[25, 150]])

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes, 0], train_input[indexes, 1], marker='D')  # marker='D'로 지정하면 산점도를 마름모로 그림
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(train_input[indexes])
print(train_target[indexes])
print(distances)
```

### 기준을 다시 맞춰보자!

x축은 범위가 좁고(10~40), y축은 범위가 넓음(0~1000)

-> y축으로 조금만 멀어져도 거리가 아주 큰 값으로 계산됨

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlim((0, 1000))  # x축 범위 지정
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 데이터 전처리

두 특성(길이와 무게)의 값이 놓인 범위가 매우 다름 = 두 특성의 '스케일'이 다름

데이터를 표현하는 기준이 다르면 알고리즘이 올바르게 예측 못함 (거리 기반)

특성 값을 일정한 기준으로 맞춰 주는 작업을 '데이터 전처리'라고 함

* 표준점수 : 각 특성값이 평균에서 표준편차의 몇 배만큼 떨어져 있는지 나타냄




```python
mean = np.mean(train_input, axis=0)  # 평균 계산
std = np.std(train_input, axis=0)  # 표준 편차 계산

print(mean, std)

train_scaled = (train_input - mean) / std  # 표준 점수 계산
```

### 전처리 데이터로 모델 훈련하기

```python
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
new = ([25, 150] - mean) / std
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.fit(train_scaled, train_target)
test_scaled = (test_input - mean) / std
kn.score(test_scaled, test_target)
print(kn.predict([new]))

distances, indexes = kn.kneighbors([new])
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes,0], train_scaled[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
